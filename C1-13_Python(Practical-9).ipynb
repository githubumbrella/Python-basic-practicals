##### In order to do that, let&#39;s apply the understanding of web scraping on the Amazon website to fetch prices and technical specifications data from the web page. This Amazon link https://www.amazon.in/s?k=samsung&amp;rh=n%3A1389401031&amp;ref=nb_sb_noss contains information about some features of Samsung phones, including technical specifications, prices, discounts and rating, etc. 

```python
# Import useful libraries and classes.
from urllib.request import urlopen as uReq
from bs4 import BeautifulSoup as soup
```

```python
#html upload
my_url="https://www.amazon.in/s?k=samsung&amp;rh=n%3A1389401031&amp;ref=nb_sb_noss"
uClient= uReq(my_url)
page_html= uClient.read()
uClient.close()
```

```python
my_url
```

```python
uClient
```

```python
page_html
```

```python
page_soup= soup(page_html)
page_soup
```

```python
#read class from web page.
containers= page_soup.findAll("div", {"class": "s-result-item"})
containers
```

```python
print(len(containers))
name2=[]
prices2=[]
specs2=[]
discount2=[]
rating2=[]
```

```python
for item in containers:
    name = item.find("span", class_="a-size-medium a-color-base a-text-normal")
    if name:
        print("Name:", name.text)
        name2.append(name.text)
    else:
        continue
    price = item.find("span", class_="a-offscreen")
    if price:
        prices2.append(price.text)
        print("Price:", price.text)
    else:
        prices2.append("Not available")
        print("Price: Not available")
    specs = item.find("span", class_="a-size-base-plus")
    if specs:
        specs2.append(specs.text)
        print("Specs:", specs.text)
    else:
        specs2.append("Not available")
        print("Specs: Not available")
    discount = item.find("span", class_="a-letter-space")
    if discount:
        discount2.append(discount.text)
        print("Discount:", discount.text)
    else:
        discount2.append("Not available")
        print("Discount: Not available")
    rating = item.find("span", class_="a-icon-alt")
    if rating:
        rating2.append(rating.text)
        print("Rating:", rating.text)
    else:
        rating2.append("Not available")
        print("Rating: Not available")
    print("\n")
```

```python
import pandas as pd
df1 = pd.DataFrame({'Product Name':name2,'Prices':prices2,'Ratings':rating2,'spects':specs2,'Discount':discount2})
df1
```

```python
df1.to_csv('data2.csv')
data = pd.read_csv("data2.csv", encoding="latin-1")
data
```
