#The show() function in pyplot module of matplotlib library is used to display all figures.

```python
#The matplotlib.pyplot.plot() function provides a unified interface 
#for creating different types of plots. 
#Example: Draw a line in a diagram from position (0,0) to position (6,250):
#plot() function to plot values as x,y coordinates in a data plot. 
# In this case, plot() takes 2 parameters for specifying plot coordinates: 
# Parameter for an array of X axis coordinates.
# Parameter for an array of Y axis coordinates.

import numpy as np
xpoints = np.array([0, 6])
ypoints = np.array([0, 250])     #The x-axis is the horizontal axis.The y-axis is the vertical axis.
plt.plot(xpoints, ypoints)
plt.show()
```

```python
#To plot only the markers, you can use shortcut string notation parameter 'o', which means 'rings'.
#Example: Draw two points in the diagram, one at position (1, 3) and one in position (8, 10):
xpoints = np.array([1, 8])
ypoints = np.array([3, 10])
plt.plot(xpoints, ypoints,'o')
plt.show()
```

```python
#Multiple Points
#Draw a line in a diagram from position (1, 3) to (2, 8) then to (6, 1) and finally to position (8, 10):
xpoints = np.array([1, 2, 6, 8])
ypoints = np.array([3, 8, 1, 10])
plt.plot(xpoints, ypoints, marker='*')
plt.show()
```

```python
#Set Font Properties for Title and Labels
#You can use the fontdict parameter in xlabel(), ylabel(), and title() to set font properties for the title and labels.

x = np.array([80, 85, 90, 95, 100, 105, 110, 115, 120, 125])
y = np.array([240, 250, 260, 270, 280, 290, 300, 310, 320, 330])

font1 = {'family':'normal','color':'blue','size':25}
font2 = {'family':'serif','color':'darkred','size':15}

plt.title("Sports Watch Data", fontdict = font1)
plt.xlabel("Average Pulse", fontdict = font2)
plt.ylabel("Calorie Burnage", fontdict = font2)
plt.plot(x, y)
plt.show()
```

```python
#Position the Title
#You can use the loc parameter in title() to position the title.
#Legal values are: 'left', 'right', and 'center'. Default value is 'center'.

x = np.array([80, 85, 90, 95, 100, 105, 110, 115, 120, 125])
y = np.array([240, 250, 260, 270, 280, 290, 300, 310, 320, 330])
plt.title("Sports Watch Data", loc = 'centre', fontdict = font2)
plt.xlabel("Average Pulse",font1)
plt.ylabel("Calorie Burnage", font1)
plt.plot(x, y)
plt.show()
```

```python
#You can specify the colormap with the keyword argument cmap with the value of the colormap, 
#in this case 'viridis' which is one of the built-in colormaps available in Matplotlib.
#In addition you have to create an array with values (from 0 to 100), one value for each of the point in the scatter plot:
#Create a color array, and specify a colormap in the scatter plot:

x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])
colors = np.array([0, 10, 20, 30, 40, 45, 50, 55, 60, 70, 80, 90, 100])
plt.scatter(x, y, c=colors, cmap='viridis')
plt.show()
```

Histogram
A histogram is a graph showing frequency distributions.
It is a graph showing the number of observations within each given interval.
Example: Say you ask for the height of 250 people, the  histogram look like this:
2 people from 140 to 145cm
5 people from 145 to 150cm
etc.

The hist() function will use an array of numbers to create a histogram, 
the array is sent into the function as an argument.
For simplicity we use NumPy to randomly generate an array with 250 values,
where the values will concentrate around 170, 
and the standard deviation is 10. 

```python
import numpy as np
x = np.random.normal(170, 10, 100)
print(x)
```

```python
x = np.random.normal(170, 10, 100)
plt.hist(x)
plt.show() 
```

```python
#Creating Pie Charts
#With Pyplot, you can use the pie() function to draw pie charts:
```

```python
y = np.array([35, 25, 25, 15, 6, 1])
plt.pie(y)
plt.show() 
```

```python
#Add labels to the pie chart with the label parameter.
#The label parameter must be an array with one label for each wedge:
y = np.array([35, 25, 25, 15])
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]
plt.pie(y, labels = mylabels)
plt.show() 
```

```python
#The startangle parameter is defined with an angle in degrees, default angle is 0:

y = np.array([35, 25, 25, 15, 5])
mylabels = ["Apples", "Bananas", "Cherries", "Dates", "orange"]

plt.pie(y, labels = mylabels, startangle = 100, shadow=True)
plt.show() 
```

```python
import matplotlib.pyplot as plt
import numpy as np

y = np.array([35, 25, 25, 15])
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]
myexplode = [0, 0.2, 0, 0]
mycolors = ["black", "hotpink", "b", "green"]

#plt.pie(y, labels = mylabels, explode = myexplode, shadow = True)
#plt.pie(y, labels = mylabels)
#plt.legend(title = "MY Fruits:")

plt.pie(y, explode=myexplode, colors=mycolors,
        autopct='%1.1f%%', shadow=True, startangle=90, radius=0.9,
        center=(0.5,0.5),
        frame=False, pctdistance=1.125)
plt.show() 
```
